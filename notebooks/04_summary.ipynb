```python
# Summary and Conclusions

## Project Overview

**Objective:** Analyze whether news sentiment can predict stock price movements for MSFT, TSLA, and VISA.

**Approach:**
1. Collected 30 days of financial news from News API
2. Analyzed sentiment using OpenAI GPT-3.5
3. Merged with stock price data from Yahoo Finance
4. Calculated correlations between sentiment and price movements

## Key Findings

### Data Collected
- **244 company-specific articles** across MSFT (60% coverage), TSLA (80% coverage), VISA (90% coverage)
- **100 macro-economic articles** (35% coverage)
- **Date range:** January 26 - February 23, 2026

### Sentiment Distribution
- Company news: Slightly positive overall (avg: 0.18)
- Macro news: Neutral (avg: 0.05)

### Correlation Results

**Same-Day Correlation:**
- MSFT: +0.14 (weak positive)
- TSLA: -0.01 (no relationship)
- VISA: +0.19 (weak positive)

**Next-Day Correlation:**
- MSFT: -0.12 (weak negative)
- TSLA: -0.33 (moderate negative - contrarian)
- VISA: +0.52 (moderate positive - **strongest predictor**)

## Limitations

1. **Sample Size:** Only 30 days of data (12-18 days with news per stock)
2. **Timing Mismatch:** Daily closing prices don't capture intraday reactions to news
3. **Data Granularity:** Missing after-hours news impact (earnings typically released after 4pm)
4. **Equal Weighting:** All articles treated equally (minor news vs major events)

## Recommendations for Future Work

1. **Extend time period:** 3-6 months minimum for statistical significance
2. **Use intraday data:** Capture immediate market reaction to news
3. **Weight by importance:** Focus on major events (earnings, acquisitions, CEO changes)
4. **Include social media:** Twitter/Reddit sentiment for retail investor behavior
5. **Compare opening prices:** Better alignment with after-hours news

## Technical Stack

- **Data Collection:** News API, yfinance
- **AI/NLP:** OpenAI GPT-3.5 Turbo
- **Analysis:** Pandas, NumPy
- **Visualization:** Matplotlib, Seaborn

## Conclusion

While the current analysis shows weak correlation between news sentiment and stock price movements, the project successfully demonstrates:
- End-to-end AI pipeline development
- API integration and data collection
- NLP sentiment analysis application
- Financial data analysis methodology
- Critical evaluation of results and limitations

The weak correlation is primarily due to sample size and timing constraints rather than the fundamental hypothesis being invalid. With longer time periods and intraday pricing data, sentiment analysis could prove to be a valuable predictive tool.
```
